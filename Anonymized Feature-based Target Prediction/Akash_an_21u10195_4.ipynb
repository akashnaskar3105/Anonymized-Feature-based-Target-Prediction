```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/Train_Data.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.fillna(df.median(),inplace=True)
df.shape
```

```python
X_train = df.drop(columns=["pred"])
y_train = df.pred
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
X_train.pc=label_encoder.fit_transform(X_train.pc)
X_train.ma=label_encoder.fit_transform(X_train.ma)
```

```python
DataFrame = pd.read_csv('/content/Test_Data.csv')
DataFrame.head()
```

```python
X_test =DataFrame.drop(columns=["index"])
y_test=DataFrame.index
```

```python
DataFrame.isna().sum()
```

```python
label_encoder=LabelEncoder()
X_test.pc=label_encoder.fit_transform(X_test.pc)
X_test.ma=label_encoder.fit_transform(X_test.ma)
```

```python
model=MultinomialNB()
```

```python
 model.fit(X_train, y_train)
 y_pred = model.predict(X_test)
```

```python
pred=pd.DataFrame(y_pred)
pred.columns=["pred"]
pred.to_csv("Sample_Submission.csv",index=False)
```
